```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x=house[['Area','Bedrooms','Bathrooms','Floors','YearBuilt']]
y=house['Price']
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plot_data = pd.DataFrame({'Area': X_test['Area'], 'Actual_Price': y_test, 'Predicted_Price': y_pred})
plot_data_sorted = plot_data.sort_values(by='Area')

plt.scatter(plot_data_sorted['Area'], plot_data_sorted['Actual_Price'], color='blue', label='Actual Data')
plt.plot(plot_data_sorted['Area'], plot_data_sorted['Predicted_Price'], color='red', label='Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.title('House Price Prediction Using Linear Regression (Area vs. Price)')
plt.show()
```
