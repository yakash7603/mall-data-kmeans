```python
import pandas as pd
import numpy as np
```

```python
malldata=pd.read_csv('/content/Mall_Customers.csv')
malldata
```

```python
malldata.drop("Gender",axis=1,inplace=True)
malldata
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X=malldata[['Annual Income (k$)','Spending Score (1-100)']]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
kmeans=KMeans(n_clusters=5,random_state=42)
clusters=kmeans.fit_predict(X_scaled)
malldata['Cluster']=clusters
```

```python
centroids=kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters,cmap='rainbow',s=60)
plt.xlabel('Annual Income (Scaled)')
plt.ylabel('Spending Score (Scaled)')
plt.title("customer segments usimg kmeams clustering")
plt.show()
```
